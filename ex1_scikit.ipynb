```
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```
def normalize(X):
  for i in range(X.shape[1]):
    a = X[:,i]
    m = np.mean(a)
    s = np.std(a)
    a = (1.*a-m)/s
    X[:,i] = a
  return X
```

```
data = pd.read_csv('/content/ex1data2.txt')
data = data.to_numpy()
X = data[:,0:2]
Y = data[:,2]
X = X.astype('float64')
Y = Y.astype('float64')
print(np.mean(X[:,0]),np.mean(X[:,1]),np.std(X[:,0]),np.std(X[:,1]))
X = normalize(X)
X = np.hstack((np.ones((X.shape[0],1)),X))
print(X)
```

```
reg = LinearRegression().fit(X,Y)
theta = reg.coef_
print(np.dot(np.array([[1,(1650-1998.4347826086957)/(794.5531599632346),(3-3.1739130434782608)/(0.7605589428059087)]]),theta))
print(reg.coef_,reg.intercept_)
print(reg.predict(np.array([[1,(1650-1998.4347826086957)/(794.5531599632346),(3-3.1739130434782608)/(0.7605589428059087)]])))
```
